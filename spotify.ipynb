```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

spotify_1 = pd.read_csv('/content/dataset.csv')
print(spotify_1.head(10))

#1)Data Understanding & Exploration:

#What are the columns in the dataset, and what do they represent?
columns_spot = spotify_1.columns
print(columns_spot)

#Are there any missing values or duplicates that need to be handled?
missing_values = spotify_1.isnull()
print('no missing values')

#What is the distribution of numerical features like duration, tempo, and loudness?
#duration:
duration_x = spotify_1['duration_ms'].agg([np.mean, np.median, np.std, np.min, np.max])
print(duration_x)
plt.figure(figsize=(10, 6))
plt.hist(spotify_1['duration_ms'], bins=30, alpha=0.7, color='skyblue', edgecolor='black')  # Use original data for the histogram
plt.title('Distribution Of Duration (ms)')
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.show()

#tempo:
tempo_x = spotify_1['tempo'].agg([np.mean, np.median, np.std, np.min, np.max])
print(tempo_x)
plt.figure(figsize=(10, 6))
plt.hist(spotify_1['tempo'], bins=30, alpha=0.7, color='lightgreen', edgecolor='black')  # Use original data for the histogram
plt.title('Distribution of Tempo (BPM)')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Frequency')
plt.show()

#loudness:
loudness_x = spotify_1['loudness'].agg([np.mean, np.median, np.std, np.min, np.max])
print(loudness_x)
plt.figure(figsize=(10, 6))
plt.hist(spotify_1['loudness'], bins=30, alpha=0.7, color='salmon', edgecolor='black')  # Use original data for the histogram
plt.title('Distribution Of Loudness (dB)')
plt.xlabel('Loudness (dB)')
plt.ylabel('Frequency')
plt.show()

# 2)Popularity & Trends Analysis:

# What are the most popular songs in the dataset?
pop_song = spotify_1.loc[spotify_1['popularity'].max()]
print('popular songs:', pop_song)

# Get the top 10 popular songs
top_songs = spotify_1.nlargest(10, 'popularity')
print("Top 10 Popular Songs:")
print(top_songs)

# Which artists appear most frequently in the dataset?
artist_counts = spotify_1['artists'].value_counts()
most_freq_artist = artist_counts.max()
print(artist_counts)
print(most_freq_artist)

# How has the popularity of different genres changed over time?
diff_generes = spotify_1['track_genre'].max()
print(diff_generes)

# What is the average duration of songs over the years?
avg_duration = spotify_1['duration_ms'].max()
print(avg_duration)

# 3)Audio Features Analysis:

# What is the relationship between danceability and energy?
dance_energy = spotify_1['danceability'].corr(spotify_1['energy'])
print('the relationship between danceability and enegry:', dance_energy)

threshold = spotify_1['danceability'].median()
colors = np.where(spotify_1['danceability'] > threshold, 'skyblue', 'pink')
plt.scatter(spotify_1['danceability'], spotify_1['energy'], alpha= 0.6, color=colors)
plt.title('The Relationship Between Danceability And Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()

# Do songs with higher loudness tend to be more popular?
high_pop = spotify_1['loudness'].corr(spotify_1['popularity'])
print('Correlation between loudness and popularity:', high_pop)

# Detect outliers using the IQR method
Q1 = spotify_1['loudness'].quantile(0.25)
Q3 = spotify_1['loudness'].quantile(0.75)
IQR = Q3 - Q1
outliers = (spotify_1['loudness'] < (Q1 - 1.5 * IQR)) | (spotify_1['loudness'] > (Q3 + 1.5 * IQR))

# Color outliers differently
colors = np.where(outliers, 'red', 'black')

# Scatter plot
plt.scatter(spotify_1['loudness'], spotify_1['popularity'], alpha=0.6, color=colors)

# Trend line
z = np.polyfit(spotify_1['loudness'], spotify_1['popularity'], 1)
p = np.poly1d(z)
plt.plot(spotify_1['loudness'], p(spotify_1['loudness']), color='blue', linewidth=2, label='Trend Line')

# Labels and title
plt.title('The Correlation Between Loudness and Popularity')
plt.xlabel('Loudness')
plt.ylabel('Popularity')
plt.legend()
plt.show()

# Are there any differences in tempo across different genres?
diff_tempo = spotify_1.groupby('track_genre')['tempo'].describe()
print(diff_tempo)

diff_tempo.plot(kind='bar', figsize=(30, 6))
plt.title('Tempo Differences Across Genres')
plt.xlabel('Genre')
plt.ylabel('Tempo')
plt.show()

# Do sadder songs (lower valence) tend to be more popular?
sad_songs = spotify_1['valence'].corr(spotify_1['popularity'])
print('Correlation between valence and popularity:', sad_songs)

threshold = spotify_1['valence'].median()
colors = np.where(spotify_1['valence'] > threshold, 'green', 'Teal')

#scatter plot
plt.scatter(spotify_1['valence'], spotify_1['popularity'], alpha= 0.6, color=colors)
plt.title('The Correlation Between valence And popularity')
plt.xlabel('Valence')
plt.ylabel('Popularity')
plt.show()


# 4)Genre-Based Analysis:

# Which genres are the most common in the dataset?
#counts
genres_counts = spotify_1['track_genre'].value_counts()
print(genres_counts)

unique_genres = spotify_1['track_genre'].unique()
print(unique_genres)

#sorted
sorted_genres = genres_counts.sort_values(ascending = False)
print(sorted_genres)
#visualize
spotify_1['track_genre'].value_counts().plot(kind='bar', figsize=(30, 6))
plt.title('Most Common Genres in the Dataset')
plt.xlabel('Genres')
plt.ylabel('Number of Songs')
plt.show()

#What are the top genres based on popularity?
group_genres = spotify_1.groupby('track_genre')['popularity']
mean_genres = group_genres.mean()
sort_genres = mean_genres.sort_values(ascending=False)
print(sorted_genres)
#plot
sorted_genres.plot(kind = 'bar',figsize=(20, 6), color = 'purple')
plt.title('Genres Based On Popularity')
plt.xlabel('Genres')
plt.ylabel("Popularity")

# Are some genres consistently more energetic or danceable than others?
mean_values = spotify_1.groupby('track_genre')[['energy', 'danceability']].mean()
print(mean_values)
sort_x = mean_values.sort_values(by = ['energy', 'danceability'], ascending = False)
print(sort_x)
#plot
sort_x.plot(kind = 'bar',figsize=(20, 6))
plt.title('Some Genres That are More Energetic And Danceable Than Others')
plt.xlabel('Genres')
plt.ylabel('Energy and Danceability')

# 5)Yearly Trends & Evolution:

# Is there a trend in loudness, danceability, or valence over time?
mean_x= spotify_1.groupby('track_genre')[['energy', 'danceability', 'valence']].mean()
print(mean_x)
sort_y = mean_x.sort_values(by = ['energy', 'danceability','valence'],ascending = False)
print(sort_y)
#plot
sort_y.plot(kind = 'bar',figsize = (20,6))
plt.title('Comparative Analysis of Energy, Danceability, and Valence Across Genres')
plt.xlabel('Genres')
plt.ylabel('Mean Values of Energy, Danceability, and Valence')
```
